```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import joblib
```

```python
start_data = pd.read_excel('data_with_point.xlsx')
start_data.head()
```

## Подготовка исходного датасета

```python
mas_of_letters = []
for row in start_data.iterrows():
    if row[1][8] == 'Т':
        mas_of_letters.append(1)
    elif row[1][8] == 'С':
        mas_of_letters.append(2)
    else:
        mas_of_letters.append(3)
start_data['J'] = mas_of_letters
start_data.describe()
```

```python
def standartization(mas):
    mean = np.average(mas)
    new_mas = []
    for el in mas:
         new_mas.append(abs(el - mean))
    return new_mas
    
```

## Обучения модели на данных без предобработки

```python
def learning_fun(data_set, name = None):
    target = start_data['Points']
    features = data_set
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=10)
    model = LinearRegression()
    model.fit(X_train, y_train)
    if model.score(X_train, y_train) > 0.71:
        print(f'------------{name}----------')
        print('Training data r-squared:',round(model.score(X_train, y_train),4))
        print('Test data r-squared:', round(model.score(X_test, y_test),4))

        print('Intercept', model.intercept_)
#     print(pd.DataFrame(data=model.coef_,index =X_train.columns,columns=['coef_points']))
    
        output = model.predict(X_test)
        output_data = pd.DataFrame(data=X_test)
        output_data['Predict'] = output
        print(output_data)
        return model
    else:
        return None
```

```python
def handler_func(mas_phonk):
    letter_mas = ['A','B','C','D','E','F','G','J']
    for name, func in mas_phonk.items():
        for letter in letter_mas:
            in_process_data = start_data.copy()
            in_process_data[letter] = func(in_process_data[letter])
            learning_fun(in_process_data,f'{name}---{letter}')
```

```python
dict_phonk = {'Тангес': np.tan,
              'Синус' : np.sin,
              'Косинус': np.cos,
              'Корень' : np.sqrt,
              'Стандартизация' : standartization
}
handler_func(dict_phonk)
```

```python
def normal_func(mas):
    pass
    return mas
```

```python
def handler_func_update(mas_phonk):
    model_mas = {}
    for name, val in mas_phonk.items():
        in_process_data = start_data.drop('Points', axis=1)
        for letter in val[1]:
            in_process_data[letter] = val[0](in_process_data[letter])
        model_mas[name] = learning_fun(in_process_data,f'{name}---{letter}')
    return model_mas
```

```python
dict_phonk = {'Тангес': [np.tan, ['D','F','G']],
              'Синус' : [np.sin, ['D','G','J']],
              'Косинус': [np.cos, ['D','G','J']],
              'Корень' : [np.sqrt, ['B','E','G']],
              'Стандартизация' : [standartization, ['F','G','J']],
              'Без обработки' : [normal_func, ['F']]
}
result = handler_func_update(dict_phonk)
```

```python
for name, model in result.items():
    # save the model to disk
    if model == None:
        continue
    filename = f'{name}.joblib'
    joblib.dump(model, filename)
```
